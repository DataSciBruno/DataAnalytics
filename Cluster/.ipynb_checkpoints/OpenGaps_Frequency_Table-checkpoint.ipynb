```python
import yfinance as yf
import pandas as pd
import plotly.express as px
from scipy import stats
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
```

```python
ticker = input('Type in the ticker of the stock: ')
```

```python
start_date = input('Start date (YYYY-MM-DD): ')
end_date = input('End date (YYYY-MM-DD): ')
```

```python
data = yf.download(ticker,start_date,end_date)
data
```

```python
data['Gap'] = (data['Open'] - data['Close'].shift(1)) / data['Close'].shift(1) * 100
data['Pct_Change'] = data['Close'].pct_change() * 100
data
```

```python
data=data.dropna()
data
```

```python
data['Close_vs_Open'] = (data['Close']/ data['Open']* 100)-100
data
```

```python
thresholds = [-np.inf, -10, -7.5, -5, -2.5, 0, 2.5, 5, 7.5, 10, np.inf]
labels = ['<-10', '-10 < -7.5', '-7.5 < -5', '-5 < -2.5', '-2.5 < 0', '0 > 2.5', '2.5 > 5', '5 > 7.5', '7.5 > 10', '>10']

# Create the frequency class column
data['FrequencyClass'] = pd.cut(data['Gap'], bins=thresholds, labels=labels, right=False)
data
```

```python
frequency_data = data['FrequencyClass'].value_counts().reset_index()
frequency_data.columns = ['Class', 'Frequency']
```

```python
frequency_data = frequency_data.sort_values('Class')
```

```python
frequency_data=frequency_data.set_index('Class')
```

```python
frequency_data
```

```python
median =[]
mean=[]
for label in labels:
    subset = data[data['FrequencyClass'] == label]
    mean_value = subset['Pct_Change'].mean()
    median_value = subset['Pct_Change'].median()
    median.append(round(median_value,2))
    mean.append(round(mean_value,2))
```

```python
frequency_data['Pct_Change Median'] = median
frequency_data['Pct_Change Mean'] = mean
frequency_data
```
