```python
from tika import parser

input_pdf_path = '01_2022.pdf'
nome_arq_exp= input_pdf_path[:-4]

raw = parser.from_file(input_pdf_path)
resultList = list(raw.items())
text = resultList[1][1]
text2 = text.split('\n')
prefix = '1-BOVESPA'
trades =[]
for i in text2:
    if prefix in i:
        trades.append(i)
trades
```

```python
op_opcoes=[]
for i in trades:
    if "1-BOVESPA V VISTA" in i or "1-BOVESPA C VISTA"  in i:
        None
    else:
        op_opcoes.append(i)
op_opcoes
```

```python
import pandas as pd
df_trades = pd.DataFrame(op_opcoes)
df_trades=df_trades[0].str.split(' ', expand=True)
df_trades
```

```python
try:
    df_trades_limpo=df_trades[[0,1,5,6,16,17,23,24,25,26,27,28]]
except:
    df_trades[28]=None
    df_trades_limpo=df_trades[[0,1,5,6,16,17,23,24,25,26,27,28]]
df_trades_limpo
```

```python
lista_tr = []

for i in range(len(df_trades_limpo)):    
    if df_trades_limpo[28][i] is None:
        bov=df_trades_limpo[0][i]
        op=df_trades_limpo[1][i]
        data=df_trades_limpo[5][i]
        ticker=df_trades_limpo[6][i]
        qtd=df_trades_limpo[24][i]
        vu=df_trades_limpo[25][i]
        total=df_trades_limpo[26][i]
        lista_tr.append((bov,op,data,ticker,qtd,vu,total))
    elif df_trades_limpo[28][i] is "C" or df_trades_limpo[28][i] is "D":
        bov=df_trades_limpo[0][i]
        op=df_trades_limpo[1][i]
        data=df_trades_limpo[5][i]
        ticker=df_trades_limpo[6][i]
        qtd=df_trades_limpo[25][i]
        vu=df_trades_limpo[26][i]
        total=df_trades_limpo[27][i]
        lista_tr.append((bov,op,data,ticker,qtd,vu,total))
    else:
        bov=df_trades_limpo[0][i]
        op=df_trades_limpo[1][i]
        data=df_trades_limpo[5][i]
        ticker=df_trades_limpo[6][i]
        qtd=df_trades_limpo[26][i]
        vu=df_trades_limpo[27][i]
        total=df_trades_limpo[28][i]
        lista_tr.append((bov,op,data,ticker,qtd,vu,total))

lista_tr 
```

```python
df_lista_trades = pd.DataFrame(lista_tr)
df_lista_trades
```

```python
df_lista_trades[4] = df_lista_trades[4].str.replace('.', '').astype(int)
df_lista_trades[5] = df_lista_trades[5].str.replace(',', '.').astype(float)
df_lista_trades[6] =df_lista_trades[6].str.replace('.', '').str.replace(',', '.').astype(float)

df_agrupado =df_lista_trades.groupby([0, 1, 2, 3, 5]).agg({4: 'sum', 6: 'sum'}).reset_index()
df_agrupado[2]=f"{nome_arq_exp}"
df_agrupado
```

```python
df_agrupado.to_excel(f"{nome_arq_exp}.xlsx")
```

```python
import os
import pandas as pd

# Pasta contendo os arquivos XLSX
pasta = "C:\\Users\\bviei\\BRUNO\\IRPF\\IRPF 2022-2023\\BRUNO\\DONE XP\\NOTAS"

# Lista para armazenar os DataFrames dos arquivos
dataframes = []

# Percorrer todos os arquivos XLSX na pasta
for arquivo in os.listdir(pasta):
    if arquivo.endswith('.xlsx'):
        caminho_arquivo = os.path.join(pasta, arquivo)
        
        # Ler o arquivo XLSX e adicionar o DataFrame à lista
        df = pd.read_excel(caminho_arquivo)
        dataframes.append(df)

# Concatenar os DataFrames em um único DataFrame
df_final = pd.concat(dataframes, ignore_index=True)

# Exportar o DataFrame final para um arquivo XLSX
caminho_saida = 'caminho_do_arquivo_de_saida.xlsx'
df_final.to_excel(caminho_saida, index=False)

print("Arquivo unificado criado com sucesso!")
```

```python
df_final.drop("Unnamed: 0", axis=1, inplace=True)
df_final
```

```python
df_final.sort_values(by=[3, 2], inplace=True)
df=df_final
```

```python
# Criar nova coluna
df['Operação Encerrada'] = ''

# Agrupar por ativo e verificar se a quantidade de C e V é igual
grupo = df.groupby([3])[4].value_counts()
ativos_encerrados = grupo[grupo % 2 == 0].index.get_level_values(0)

# Verificar se cada linha representa uma operação encerrada
for index, row in df.iterrows():
    if row[3] in ativos_encerrados:
        df.at[index, 'Operação Encerrada'] = 'Sim'
    else:
        df.at[index, 'Operação Encerrada'] = 'Não'
```

```python
df_operacoes_abertas = df[df['Operação Encerrada'] == 'Não']
```

```python
df_operacoes_encerradas = df[df['Operação Encerrada'] == 'Sim']
df_operacoes_encerradas
```

```python
df_operacoes_encerradas = df_operacoes_encerradas.reset_index(drop=True)
```
