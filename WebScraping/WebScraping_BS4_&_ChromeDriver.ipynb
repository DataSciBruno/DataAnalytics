```python
import requests, time, random
from bs4 import BeautifulSoup

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

import time
from time import sleep

import pandas as pd
```

```python
url = 'https://www.vagas.com.br/mapa-de-carreiras/cargos'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')
```

```python
positions=[]

for link in soup.find_all('a'):
    position=link.get('href')
    position =str(position)
    position = position[7:] #eliminating the 'cargos/' before the position
    positions.append(position)
positions
```

```python
positions_example = positions[190:200]
positions_example
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
details_positions_examples=[]

for i in positions_example:
    try:
        postion_name=i
        postion_name=postion_name.replace('-',' ') #removing '-' from the position
        postion_name=postion_name.capitalize() #capitalizing the first letter of the words
        url = 'https://www.vagas.com.br/mapa-de-carreiras/cargos/'+i
        driver.get(url)
        src= driver.page_source
        soup = BeautifulSoup(src,'lxml')
        time.sleep(5)
        details = soup.find(class_="tag")
        details= details.get_text()
        details= details.split('\n')
        details = list(filter(None, details))
        details_positions_examples.append((postion_name,details))
    except:
        None
```

```python
df = pd.DataFrame(details_positions_examples)
df.rename(columns = {0:'Postion', 1:'Details'}, inplace = True)
df
```
