# LIBRARIES

```python
import yfinance as yf
import pandas as pd
import plotly.express as px
from scipy import stats
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
```

# INPUT TICKER

```python
ticker = input('Type in the ticker of the stock: ')
```

# INPUT START & END DATES TO DOWNLOAD DATA

```python
start_date = input('Start date (YYYY-MM-DD): ')
end_date = input('End date (YYYY-MM-DD): ')
```

# DOWNLOADING DATA

```python
data = yf.download(ticker,start_date,end_date)
data
```

# PREPROCESSING

```python
data['Pct_Change'] = data['Adj Close'].pct_change()*100
```

```python
data.dropna(inplace=True)
```

```python
data.reset_index(inplace=True)
```

# PLOT

```python
data
```

```python
plt.scatter(data['Date'], data['Pct_Change'])
plt.xlabel('Date')
plt.ylabel('Pct_Change')
plt.title('Scatter Plot')
plt.show()
```

# REMOVING OUTLIERS

```python
outliers = np.array(data.Pct_Change)

z_scores = stats.zscore(outliers)

threshold = 3.0


outliers_list = np.where(np.abs(z_scores) > threshold)

outliers_list=outliers_list[0]
outliers_list
```

```python
data_no_outliers = data.reset_index(drop=True)
data_no_outliers = data_no_outliers.drop(outliers_list)
data_no_outliers
```

```python
plt.scatter(data_no_outliers['Date'], data_no_outliers['Pct_Change'])
plt.xlabel('Date')
plt.ylabel('Pct_Change')
plt.title('Scatter Plot')
plt.show()
```
